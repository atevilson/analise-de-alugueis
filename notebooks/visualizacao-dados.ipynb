```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [4.0, 3.0]
plt.style.use('classic')
plt.rcParams['figure.dpi'] = 100
plt.rcParams['savefig.dpi'] = 120
plt.rcParams['legend.fontsize'] = 'small'
plt.rcParams['figure.titlesize'] = 'medium'
plt.rcParams['font.size'] = 9
plt.rcParams['axes.ymargin'] = 1
plt.rcParams['axes.xmargin'] = 0
```

```python
dado = pd.read_csv("imoveis-residenciais.csv", sep= ";")
```

```python
dado.head()
```

```python
area = plt.figure()
```

```python
g1 = area.add_subplot(2,2,1)
g2 = area.add_subplot(2,2,2)
g3 = area.add_subplot(2,2,3)
g4 = area.add_subplot(2,2,4)
```

```python
g1.scatter(dado.Valor, dado.Area)
g1.set_title("Valor por área")

g2.hist(dado.Valor)
g2.set_title("Histograma")

dados_g3 = dado.Valor.sample(100)
dados_g3.index = range(dados_g3.shape[0])
g3.plot(dados_g3)
g3.set_title("Amostras de valores")

grupo = dado.groupby("Tipo")["Valor"]
label = grupo.mean().index
valor = grupo.mean().values
g4.bar(label, valor)
g4.set_title("Valor médio por tipo")
```
